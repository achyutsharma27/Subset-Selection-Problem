```python
# Find all the subsets from a set of numbers whose sum is zero.
#Constraint: Subset size must be 5
#Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4

from itertools import combinations

numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
subset_size = 5
result = []

# Generate all combinations of subset_size from the numbers
combinations_list = list(combinations(numbers, subset_size))

# Check each combination for a sum of zero
for combination in combinations_list:
    if sum(combination) == 0:
        result.append(combination)

# Print the subsets with a sum of zero
for subset in result:
    print(subset)
```
